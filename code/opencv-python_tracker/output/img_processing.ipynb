```python
import cv2
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width + 100, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width + 100, 0))
    return dst

def get_concat_v(im1, im2):
    dst = Image.new('RGB', (im1.width, im1.height + im2.height + 100))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height + 100))
    return dst
```

```python
def get_img_list(n_start, n_end):
    img_list=[]
    for i in range(n_start, n_end+1):
        img_name = "./car1_kcf_opencv/car1_kcf_opencv"+str(int(i))+".jpg"
        img = Image.open(img_name)
        img_list.append(img)
    return img_list

def concat_h(n_start=0, n_end=100):
    img_list = get_img_list(n_start, n_end)
    res = img_list[0]
    for i in range(1, len(img_list)):
        res = get_concat_h(res, img_list[i])
    return res
```

```python
res = concat_h(n_start=1, n_end=10)
res.save('car1_kcf_opencv/car1_kcf_opencv_01-10.jpg')
plt.imshow(res)

res = concat_h(n_start=11, n_end=20)
res.save('car1_kcf_opencv/car1_kcf_opencv_11-20.jpg')
plt.imshow(res)

res = concat_h(n_start=21, n_end=30)
res.save('data/car1_opencv_21-30.jpg')
plt.imshow(res)

res = concat_h(n_start=31, n_end=40)
res.save('data/chaplin_kcf_opencv_31-40.jpg')
plt.imshow(res)

res = concat_h(n_start=41, n_end=50)
res.save('data/chaplin_kcf_opencv_41-50.jpg')
plt.imshow(res)

res = concat_h(n_start=51, n_end=60)
res.save('data/chaplin_kcf_opencv_51-60.jpg')
plt.imshow(res)

res = concat_h(n_start=61, n_end=70)
res.save('data/chaplin_kcf_opencv_61-70.jpg')
plt.imshow(res)

res = concat_h(n_start=71, n_end=80)
res.save('data/chaplin_kcf_opencv_71-80.jpg')
plt.imshow(res)

res = concat_h(n_start=81, n_end=90)
res.save('data/chaplin_kcf_opencv_81-90.jpg')
plt.imshow(res)

res = concat_h(n_start=91, n_end=100)
res.save('data/chaplin_kcf_opencv_91-100.jpg')
plt.imshow(res)
```
