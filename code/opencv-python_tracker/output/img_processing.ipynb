```python
import cv2
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width + 100, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width + 100, 0))
    return dst

def get_concat_v(im1, im2):
    dst = Image.new('RGB', (im1.width, im1.height + im2.height + 100))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height + 100))
    return dst
```

```python
def get_img_list(n_start, n_end):
    img_list=[]
    for i in range(n_start, n_end+1):
        img_name = "./car1_kcf_opencv/car1_kcf_opencv"+str(int(i))+".jpg"
        img = Image.open(img_name)
        img_list.append(img)
    return img_list

def concat_h(n_start=0, n_end=100):
    img_list = get_img_list(n_start, n_end)
    res = img_list[0]
    for i in range(1, len(img_list)):
        res = get_concat_h(res, img_list[i])
    return res
```

```python
res = concat_h(n_start=1, n_end=5)
res.save('car1_kcf_opencv/car1_kcf_opencv_01-05.jpg')
res = concat_h(n_start=6, n_end=10)
res.save('car1_kcf_opencv/car1_kcf_opencv_06-10.jpg')
res = concat_h(n_start=11, n_end=15)
res.save('car1_kcf_opencv/car1_kcf_opencv_11-15.jpg')

plt.imshow(res)
```

```python
img1 = Image.open('car1_kcf_opencv/car1_kcf_opencv_01-05.jpg')
img2 = Image.open('car1_kcf_opencv/car1_kcf_opencv_06-10.jpg')
img3 = Image.open('car1_kcf_opencv/car1_kcf_opencv_11-15.jpg')

res = get_concat_v(img1, img2) 
res = get_concat_v(res, img3) 

plt.imshow(res)
```
